```python
class Stack():

    def __init__(self):
        self.top = -1
        self.array_len = None
        self.arr = None

    def create(self, length):
        '''Pass size of the array'''
        self.array_len = length
        self.arr = [0 for _ in range(self.array_len)]
        return self.arr

    def isfull(self):
        '''Check whether the stack is full or not'''
        if self.top == self.array_len - 1:
            return 1
        return 0

    def isemt(self):
        '''Check whether the satck is empty or not'''
        if self.top == -1:
            return 1
        return 0

    def push(self):
        '''Pushing or insertting of elements to the stack'''
        if self.isfull():
            print("-------------Array is full-------------")
        else:
            self.top += 1
            self.arr[self.top] = input("Enter value: ")
        return self.arr

    def Pop(self):
        '''Popping elements from the stack at last'''
        if self.isemt():
            print("-------------Array is empty-------------")
        elif self.top != -1:
            self.arr[self.top] = 0
            self.top -= 1
            return self.arr

    def Display(self):
        '''Display the stack'''
        if self.top == -1:
            print("-------------This is an empty stack-------------")
        else:
            req = self.arr[:self.top + 1]
            print("Your stack:", *req)



#implementation of stack
stack = Stack()
arr_sze = int(input("Enter array length: "))

stk = stack.create(arr_sze)
cnd = True
while cnd:
    print("1. Push\n2. Pop\n3. Display\n4. Quit")
    query = int(input("Query: "))
    if query == 1:
    	stk = stack.push()

    elif query == 2:
    	stk = stack.Pop()

    elif query == 3:
        stack.Display()
    
    elif query == 4:
        print("Quitting...")
        cnd = False
    
    else:
        print("Please select valid option")
```
